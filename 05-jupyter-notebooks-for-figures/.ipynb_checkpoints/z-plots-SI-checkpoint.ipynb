#### Calculate COM along the Z axis

```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

plt.close('all')
plt.rcdefaults()
plt.rcParams['font.family'] = 'Arial'

label_fontsize  = 22
legend_fontsize = 12
tick_fontsize   = 22

output_dir = "z-figure-main-paper-19-Sep/"
os.makedirs(output_dir, exist_ok=True)

all_paths_7g8 = glob.glob("PfCRT-docked-pose/rep*7g8*/*tcl_COMz.csv")
all_paths_hb3 = glob.glob("PfCRT-docked-pose-hb3/rep*/*tcl_COMz.csv")
all_paths = all_paths_7g8 + all_paths_hb3

conds = ['gsh', 'gsd']

file_entries = []
for path in sorted(all_paths):
    parts   = path.split(os.sep)
    rep_dir = parts[1]
    rep_tok = rep_dir.split('-')
    rep     = int(rep_tok[0].lstrip('rep'))
    group   = rep_tok[1]
    cond    = rep_tok[2]
    if cond not in conds:
        continue
    file_entries.append((group, rep, cond, path))

y_mins, y_maxs = [], []
for _, _, _, path in file_entries:
    df       = pd.read_csv(path)
    y        = df.iloc[:, 1]
    smoothed = y.rolling(window=100, center=True).mean()
    y_mins.append(smoothed.min())
    y_maxs.append(smoothed.max())

global_ymin, global_ymax = -1, 29

fig, axes = plt.subplots(nrows=10, ncols=2, figsize=(10, 10))

cond_color = {
    'gsd': (1, 0, 0),
    'gsh': (0, 0, 1),
}

group_to_col = {'hb3': 0, '7g8': 1}

for group, rep, cond, path in file_entries:
    col = group_to_col[group]
    row = conds.index(cond) * 5 + (rep - 1)
    ax  = axes[row, col]

    df            = pd.read_csv(path)
    x_col, y_col  = df.columns[0], df.columns[1]
    y             = df[y_col]
    smoothed      = y.rolling(window=100, center=True).mean()
    x             = (df[x_col]/100) if pd.api.types.is_numeric_dtype(df[x_col]) else df[x_col]

    ax.plot(x, y,        color=cond_color[cond], linewidth=3, alpha=0.5)
    ax.plot(x, smoothed, color=cond_color[cond], linewidth=3, label=f"pose {rep}")

    ax.set_ylim(global_ymin, global_ymax)
    ax.legend(loc='upper left', fontsize=legend_fontsize, frameon=False)

    ax.tick_params(axis='y',
                   labelleft=(col == 0),
                   labelright=False,
                   labelsize=tick_fontsize)
    ax.tick_params(axis='x',
                   labelbottom=(row == 9),
                   labeltop=False,
                   labelsize=tick_fontsize)

    for spine in ax.spines.values():
        spine.set_edgecolor('black')
        spine.set_linewidth(2.5)

plt.tight_layout()

left, right = fig.subplotpars.left, fig.subplotpars.right
bottom, top = fig.subplotpars.bottom, fig.subplotpars.top

fig.supxlabel('Time (ns)',
              fontsize=label_fontsize,
              x=(left+right)/2, y=-0.02)
fig.supylabel('COM$_{Z}$ ()',
              fontsize=label_fontsize,
              x=-0.02, y=(bottom+top)/2)

handles = [
    Patch(color=cond_color['gsh'], label='GSH'),
    Patch(color=cond_color['gsd'], label='GSSG'),
]
fig.legend(handles=handles,
           loc='lower right',
           bbox_to_anchor=(1.0, -0.09),
           ncol=1,
           fontsize=tick_fontsize,
           frameon=False)

for ci, label in enumerate(['WT', '7G8']):
    ax = axes[0, ci]
    ax.text(0.5, 1.08, label,
            ha='center', va='bottom',
            fontsize=label_fontsize, fontweight='bold',
            transform=ax.transAxes, clip_on=False)

output_path = os.path.join(output_dir, "z_COMz_two_cols.png")
plt.savefig(output_path, dpi=300, bbox_inches='tight')
plt.show()

print(f"Saved: {output_path}")
```

#### Calculate RMSD of the protein

```python
import os
import glob
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
import numpy as np

plt.close('all')
plt.rcdefaults()
plt.rcParams['font.family'] = 'Arial'

label_fontsize = 22
legend_fontsize = 12
tick_fontsize = 22

output_dir = "z-figure-main-paper-19-Sep/"
os.makedirs(output_dir, exist_ok=True)

all_paths_7g8 = glob.glob("PfCRT-docked-pose/rep*7g8*/*prot_RMSD.csv")
all_paths_hb3 = glob.glob("PfCRT-docked-pose-hb3/rep*/*prot_RMSD.csv")
all_paths = all_paths_7g8 + all_paths_hb3

conds = ['gsh', 'gsd']

entries = []
for path in sorted(all_paths):
    parts  = path.split(os.sep)
    repdir = parts[1]
    toks   = repdir.split('-')
    rep    = int(toks[0].lstrip('rep'))
    group  = toks[1]
    cond   = toks[2]
    if cond not in conds:
        continue
    entries.append((group, rep, cond, path))

y_mins, y_maxs = [], []
for _, _, _, path in entries:
    df = pd.read_csv(path)
    x_col = df.columns[0]
    if not pd.api.types.is_numeric_dtype(df[x_col]):
        continue
    x = df[x_col] / 100
    y = df.iloc[:, 1]
    smoothed = y.rolling(window=100, center=True).mean()
    smoothed_after_50 = smoothed[x >= 50]
    if not smoothed_after_50.dropna().empty:
        y_mins.append(smoothed_after_50.min())
        y_maxs.append(smoothed_after_50.max())

if not y_mins or not y_maxs:
    print("Warning: No data found after 50 ns to determine y-limits. Exiting.")
    raise SystemExit

global_ymin = min(y_mins)
global_ymax = max(y_maxs)

fig, axes = plt.subplots(nrows=10, ncols=2, figsize=(10, 10))

cond_color = {
    'gsd': (1, 0, 0),
    'gsh': (0, 0, 1),
}

group_to_col = {'hb3': 0, '7g8': 1}

all_deltas = []

for group, rep, cond, path in entries:
    col = group_to_col[group]
    row = conds.index(cond) * 5 + (rep - 1)
    ax  = axes[row, col]

    df = pd.read_csv(path)
    x_col, y_col = df.columns[0], df.columns[1]
    if pd.api.types.is_numeric_dtype(df[x_col]):
        x = df[x_col] / 100
    else:
        x = df[x_col]
    y = df[y_col]
    smoothed = y.rolling(window=100, center=True).mean()

    ax.plot(x, y,        color=cond_color[cond], linewidth=3, alpha=0.3)
    ax.plot(x, smoothed, color=cond_color[cond], linewidth=3, label=f"pose {rep}")
    ax.set_ylim(global_ymin, global_ymax)

    mask = (x >= 50) & (x <= 200)
    sm_win = smoothed[mask]
    if not sm_win.dropna().empty:
        diff = sm_win.max() - sm_win.min()
        all_deltas.append(diff)
        ax.text(0.05, 0.9, f' = {diff:.2f} ',
                transform=ax.transAxes,
                fontsize=legend_fontsize,
                va='top', ha='left',
                bbox=dict(boxstyle='round,pad=0.2', fc='white', ec='none', alpha=0.6))

    ax.legend(loc='lower right', bbox_to_anchor=(1.0, -0.2),
              fontsize=legend_fontsize, frameon=False)

    ax.tick_params(axis='y', labelleft=(col == 0), labelright=False, labelsize=tick_fontsize)
    ax.tick_params(axis='x', labelbottom=(row == 9), labeltop=False, labelsize=tick_fontsize)

    for spine in ax.spines.values():
        spine.set_edgecolor('black')
        spine.set_linewidth(2.5)

plt.tight_layout()

left, right = fig.subplotpars.left, fig.subplotpars.right
bottom, top = fig.subplotpars.bottom, fig.subplotpars.top
fig.supxlabel('Time (ns)', fontsize=label_fontsize, x=(left + right) / 2, y=-0.02)
fig.supylabel('RMSD ()',  fontsize=label_fontsize, x=-0.05, y=(bottom + top) / 2, va='center')

handles = [
    Patch(color=cond_color['gsh'], label='GSH'),
    Patch(color=cond_color['gsd'], label='GSSG'),
]
fig.legend(handles=handles, loc='lower right', bbox_to_anchor=(1.0, -0.09),
           ncol=1, fontsize=tick_fontsize, frameon=False)

for ci, label in enumerate(['WT', '7G8']):
    ax = axes[0, ci]
    ax.text(0.5, 1.08, label,
            ha='center', va='bottom',
            fontsize=label_fontsize, fontweight='bold',
            transform=ax.transAxes, clip_on=False)

fig.subplots_adjust(left=0.04, bottom=0.05, top=0.9, right=0.95)

output_path = os.path.join(output_dir, "z_RMSD_two_cols.png")
plt.savefig(output_path, dpi=300, bbox_inches='tight')
plt.show()
print(f"Saved: {output_path}")

if all_deltas:
    average_delta = np.mean(all_deltas)
    print(f"Average  (50-200 ns): {average_delta:.2f} ")
else:
    print("No delta values were calculated.")
```

```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

plt.close('all')
plt.rcdefaults()
plt.rcParams['font.family'] = 'Arial'

total_frames  = 20000
fontsize      = 28
output_dir    = "z-figure-main-paper-18-Aug"
os.makedirs(output_dir, exist_ok=True)

three_to_one = {
    'ALA':'A','CYS':'C','ASP':'D','GLU':'E','PHE':'F','GLY':'G',
    'HIS':'H','HSP':'H','HSD':'H','HSE':'H',
    'ILE':'I','LYS':'K','LEU':'L','MET':'M','ASN':'N',
    'PRO':'P','GLN':'Q','ARG':'R','SER':'S','THR':'T','VAL':'V',
    'TRP':'W','TYR':'Y'
}
title_lig_map = {'gsd':'GSSG', 'gsh':'GSH', 'cq':'CQ'}

conds  = ['gsd','gsh']
groups = ['hb3','7g8']

all_paths_7g8 = glob.glob("PfCRT-docked-pose/rep*7g8*/z_result_tcl_protein_contacts.csv")
all_paths_hb3 = glob.glob("PfCRT-docked-pose-hb3/rep*/z_result_tcl_protein_contacts.csv")

paths = all_paths_7g8 + all_paths_hb3

entries = []
for p in sorted(paths):
    d = p.split(os.sep)[1]
    rep, grp, cond = d.split('-')[:3]
    entries.append((grp, cond, rep, p))

panel_data   = {}
panel_labels = {}
max_reps = 0
max_lbls = 0

for grp in groups:
    for cond in conds:
        key = (grp, cond)
        subset = [e for e in entries if e[0]==grp and e[1]==cond]
        data = {}
        labels_set = set()
        for _, _, rep, path in sorted(subset, key=lambda x: int(x[2].lstrip('rep'))):
            df = pd.read_csv(path)
            df['fraction'] = df['frames_in_contact'] / total_frames
            filt = df[df['fraction'] > 0.1]
            if filt.empty:
                continue
            lbls = [f"{int(r)}-{three_to_one.get(n,'X')}"
                    for r, n in zip(filt['resid'], filt['resname'])]
            fracs = filt['fraction'].values
            data[rep.lstrip('rep')] = dict(zip(lbls, fracs))
            labels_set.update(lbls)

        sorted_lbls = sorted(labels_set,
                             key=lambda x: (int(x.split('-')[0]), x.split('-')[1]))
        panel_data[key]   = data
        panel_labels[key] = sorted_lbls
        max_reps = max(max_reps, len(data))
        max_lbls = max(max_lbls, len(sorted_lbls))

fixed_width  = 0.9 / max_reps
left_global  = -fixed_width/2 - 0.1
right_global = (max_lbls - 1) + (max_reps - 1)*fixed_width + fixed_width/2 + 0.1

for grp in groups:
    fig, axs = plt.subplots(nrows=2, ncols=1,
                            figsize=(26, 20/2),
                            sharex=False)
    plt.rcParams['font.size'] = fontsize

    for ax, cond in zip(axs, conds):
        data   = panel_data[(grp, cond)]
        labels = panel_labels[(grp, cond)]
        x      = np.arange(len(labels))

        for i, (rep, vals) in enumerate(data.items()):
            heights = [vals.get(lbl, 0) for lbl in labels]
            ax.bar(x + i*fixed_width,
                   heights,
                   width=fixed_width,
                   label=f"pose {rep}")

        for spine in ax.spines.values():
            spine.set_edgecolor('black')
            spine.set_linewidth(2.5)

        mid_offset = fixed_width*(len(data)-1)/2
        for idx in range(len(labels)-1):
            sep = idx + 0.5 + mid_offset
            ax.axvline(sep, linestyle='--', linewidth=1, color='gray', zorder=0)
        end = len(labels)-0.5 + mid_offset
        ax.axvline(end, linewidth=2.5, color='black', zorder=0)

        ax.tick_params(axis='y', labelsize=fontsize)
        ax.set_xlim(left_global, right_global)
        ax.set_ylabel('Fraction of Contact', fontsize=fontsize)
        ax.set_xticks(x + mid_offset)
        ax.set_xticklabels(labels, rotation=90, fontsize=fontsize)

        display_name = 'WT' if grp.lower() == 'hb3' else grp.upper()
        ax.set_title(f"{display_name} {title_lig_map[cond]}",
                     loc='left', pad=10, fontsize=28)

    handles, labs = axs[0].get_legend_handles_labels()
    fig.legend(handles, labs,
               title='Pose',
               title_fontsize=30,
               fontsize=28,
               frameon=False,
               ncol=len(handles),
               loc='lower center',
               bbox_to_anchor=(0.5, -0.125))

    plt.tight_layout()
    plt.subplots_adjust(bottom=0.15)

    out_path = os.path.join(output_dir, f"z_contact_{grp}_no_CQ.png")
    plt.savefig(out_path, dpi=300, bbox_inches='tight')
    plt.show()
    print(f"Saved: {out_path}")
```
